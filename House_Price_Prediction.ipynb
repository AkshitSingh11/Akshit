# Importing the Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

## Importing the California Housing Dataset

```python
house_price_dataset = sklearn.datasets.fetch_california_housing()
```

```python
house_price_dataset
```

# About the dataset

Target variable is the median house value for California districts,expressed in hundreds of thousands of dollars ($100,000). This dataset was derived from the 1990 U.S. census, using one row per census block group. A block group is the smallest geographical unit for which the U.S. Census Bureau publishes sample data (a block group typically has a population of 600 to 3,000 people). A household is a group of people residing within a home. Since the average number of rooms and bedrooms in this dataset are provided per household, these columns may take surprisingly large values for block groups with few households and many empty houses, such as vacation resorts.

# Dataset to DataFrame


```python
#Loading the dataset to a Pandas dataFrame
house_price_dataframe = pd.DataFrame(house_price_dataset.data, columns=house_price_dataset.feature_names)
```

```python
# First 5 rows of our DataFrame

# MedInc- median income in block group
#HouseAge- median house age in block group
#AveRooms- average number of rooms per household
#AveBedrms- average number of bedrooms per household
#Population- block group population
#AveOccup- average number of household members
#Latitude- block group latitude
#Longitude- block group longitude

house_price_dataframe.head()
```

```python
# Adding target (price) column to the DataFrames
house_price_dataframe['price']=house_price_dataset.target
```

```python
# First 5 rows of our DataFrame with targets
house_price_dataframe.head()
```

```python
# Checking shape of the DataFrame (rows,columns)
house_price_dataframe.shape
```

```python
#Checking for missing values
house_price_dataframe.isnull().sum()
```

# Data Analysis

##Statistical measures of the dataset

```python
house_price_dataframe.describe()
```

## Understanding the correlation between various features in the dataset

```python
correlation = house_price_dataframe.corr()
```

```python
# constructing a heatmap to understand the correlation
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Greens')
```

#Splitting the data and target

```python
X = house_price_dataframe.drop(['price'], axis =1)
y= house_price_dataframe['price']
```

```python
# Print X and y
print(X)
print(y)
```

# Splitting the data into Training and Test data

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

# Model Training

## XGBoost Regressor

```python
# Loading the model
model = XGBRegressor(max_depth=6)
```

```python
# Training the model with X_train
model.fit(X_train, y_train)
```

# Evaluation

##Prediction on training data

```python
training_data_prediction = model.predict(X_train)
```

```python
print(training_data_prediction)
```

```python
#R-squared error
score_1 = metrics.r2_score(y_train, training_data_prediction)

# Mean Absolute Error
score_2 = metrics.mean_absolute_error(y_train, training_data_prediction)

print("R squared error : ", score_1)
print('Mean Absolute Error : ', score_2)
```

## Prediction on test data

```python
# Accuracy for prediction on testing data
test_data_prediction = model.predict(X_test)
```

```python
# R squared error
score_1 = metrics.r2_score(y_test, test_data_prediction)

# Mean Absolute Error
score_2 = metrics.mean_absolute_error(y_test, test_data_prediction)

print("R squared error : ", score_1)
print('Mean Absolute Error : ', score_2)
```

# Visualization

## Visualizing actual prices and predicted prices on training data

```python
plt.scatter(y_train, training_data_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Price vs Preicted Price")
plt.show()
```

## Visualizing actual prices and predicted prices on testing data

```python
plt.scatter(y_test, test_data_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Price vs Preicted Price")
plt.show()
```
